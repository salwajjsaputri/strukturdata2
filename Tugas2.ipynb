1. Double Linked List dengan 3 Node + Penyisipan Tengah + Hapus dengan Kata Kunci

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    # Menambahkan node di akhir
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # Menyisipkan node di tengah (setelah node dengan nilai tertentu)
    def insert_after(self, key, data):
        current = self.head
        while current and current.data != key:
            current = current.next
        if current is None:
            print(f"Node dengan nilai {key} tidak ditemukan")
            return
        new_node = Node(data)
        new_node.next = current.next
        new_node.prev = current
        if current.next:
            current.next.prev = new_node
        current.next = new_node

    # Menghapus node berdasarkan kata kunci
    def delete_by_key(self, key):
        current = self.head
        while current and current.data != key:
            current = current.next
        if current is None:
            print(f"Node dengan nilai {key} tidak ditemukan")
            return
        if current.prev:
            current.prev.next = current.next
        else:
            self.head = current.next
        if current.next:
            current.next.prev = current.prev
        print(f"Node dengan nilai {key} berhasil dihapus")

    # Menampilkan linked list dari head ke tail
    def display_forward(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        print(" <-> ".join(nodes))

# Membuat double linked list dengan 3 node
print("=== DOUBLE LINKED LIST ===")
dll = DoubleLinkedList()
dll.append(10)
dll.append(20)
dll.append(30)
print("Double linked list awal:")
dll.display_forward()  # Output: 10 <-> 20 <-> 30

# Penyisipan tengah (setelah node 20)
print("\nMelakukan penyisipan tengah (setelah node 20):")
dll.insert_after(20, 25)
dll.display_forward()  # Output: 10 <-> 20 <-> 25 <-> 30

# Hapus dengan kata kunci (hapus node 25)
print("\nMelakukan hapus dengan kata kunci (hapus node 25):")
dll.delete_by_key(25)
dll.display_forward()  # Output: 10 <-> 20 <-> 30
```

2. Circular Linked List dengan 3 Node + Penyisipan Akhir + Hapus Akhir

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Menambahkan node di akhir
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return
        current = self.head
        while current.next != self.head:
            current = current.next
        current.next = new_node
        new_node.next = self.head

    # Menyisipkan node di akhir (sama seperti append)
    def insert_end(self, data):
        self.append(data)

    # Menghapus node terakhir
    def delete_end(self):
        if self.head is None:
            print("Linked list kosong")
            return
        if self.head.next == self.head:
            # Hanya satu node
            print(f"Node {self.head.data} dihapus")
            self.head = None
            return
        current = self.head
        prev = None
        while current.next != self.head:
            prev = current
            current = current.next
        # current adalah node terakhir, prev adalah node sebelumnya
        prev.next = self.head
        print(f"Node terakhir ({current.data}) berhasil dihapus")

    # Menampilkan circular linked list
    def display(self):
        if self.head is None:
            print("Linked list kosong")
            return
        nodes = []
        current = self.head
        while True:
            nodes.append(str(current.data))
            current = current.next
            if current == self.head:
                break
        print(" -> ".join(nodes) + " -> (kembali ke head)")

# Membuat circular linked list dengan 3 node
print("\n=== CIRCULAR LINKED LIST ===")
cll = CircularLinkedList()
cll.append(100)
cll.append(200)
cll.append(300)
print("Circular linked list awal:")
cll.display()  # Output: 100 -> 200 -> 300 -> (kembali ke head)

# Penyisipan akhir
print("\nMelakukan penyisipan akhir (menambah node 400):")
cll.insert_end(400)
cll.display()  # Output: 100 -> 200 -> 300 -> 400 -> (kembali ke head)

# Hapus akhir
print("\nMelakukan hapus akhir:")
cll.delete_end()
cll.display()  # Output: 100 -> 200 -> 300 -> (kembali ke head)
```
